# DATA VISUALIZATION:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(y="genre",data=df,hue="genre")
plt.show()
```

# From this graph, it can be observed that video games featuring "action" as one of the genres are the most popular.

```python
plt.figure(figsize=(17,8))
sns.violinplot(x="genre",y="year",data=df,hue="genre")
```

# By this violin plot, we obtain genre popularity of video games for each year.

```python
plt.figure(figsize=(12,10))
bars=sns.countplot(y="year",data=df,hue="year")
for bar in bars.containers:
    bars.bar_label(bar)
plt.xlabel("Video Games Released")
```

# This graph illustrates that the highest number of video games were introduced in the year 2009, with a total of 1431 releases, closely followed by 2008, which saw 1428 releases.

```python
plt.figure(figsize=(20,80))
sns.countplot(y="publisher",data=df,color="green")
plt.xlabel("Video Games Released")
plt.ylabel("Publishers")
```

# Most of the video games are published by "Electronic Arts".

```python
import numpy as np
annual_sales = df.groupby('year').sum(numeric_only=True)
annual_sales = annual_sales.reset_index()
plt.figure(figsize=(10,10))
plt.plot(annual_sales['year'],annual_sales['na_sales'], label='North America Sales', color="pink",marker="o")
plt.plot(annual_sales['year'],annual_sales['eu_sales'], label='Europe Sales', color="green",marker=">")
plt.plot(annual_sales['year'],annual_sales['jp_sales'], label='Japan Sales', color="red",marker="*")
plt.plot(annual_sales['year'],annual_sales['other_sales'], label='Other Sales', marker='o')
plt.legend()
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid()
plt.xticks(np.arange(1980,2021),rotation=90)
plt.show()
```

# North America has gain significant revenue from the sales of video games, with a notable peak observed in 2008.

# CONCLUSION:
# It can be inferred that a majority of video games were published in 2009, with the "action" genre being the most popular among them. "Electronic Arts" emerges as the primary publisher of video games. North America earns the highest revenue from video game sales compared to other countries and continents.
