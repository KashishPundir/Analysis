# DATA VISUALIZATION

```python
import seaborn as sns
ax=sns.countplot(x='Gender',data=df,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)

```

# From this graph, it can be said that Males are more likely to order food online than Females

```python
sns.countplot(x="Age",data=df,hue="Age")
```

# From this graph, it can be said that Most of the customers are between the age of 22-25

```python
ax=sns.countplot(x="Marital Status",data=df,hue="Marital Status")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data=df, x='Marital Status', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

# Most of the customers are among the unmarried(single) people where males are more likely to order food online.

```python
count=sns.countplot(data=df,x="Occupation",hue="Occupation")
for bars in count.containers:
    count.bar_label(bars)
```

# From this graph, it is clearly shown that Students used to order the food online as compared to other occupations.

# CONCLUSION:

Unmarried males aged between 22 and 25, who are students(occupation), are statistically more inclined to make online food orders.
